###### groupby
In this python pandas tutorial we will learn how the groupby method can be used to group your dataset based on some criteria and then apply analytics on each of the groups. (This is similar to SQL group by.) It is also called split apply combine strategy in data science.

```python
from IPython.display import Image # To display images in jupyter notebook
import pandas as pd
df = pd.read_csv('D:\\Pandas\\CodeBasics\\datasets\\07weatherbycities.csv', sep = '\t')
df # OK so here is our dataframe
```

We want to group our data by the cities mentioned in the df and then run some analytics on the group to find the top temp, avg windspeed etc

```python
g = df.groupby("city") # City is the column that we want to groupby
g
```

```python
Image(filename = 'D:\\Pandas\\CodeBasics\\pics\\07groupbycities.jpg', height = 900, width = 900)
```

Above is a pictorial representation of the DataFrameGroupBy object mentioned above. When we pass the command df.groupby("city"), it creates the second object where you have keys and values, like a dictionary, new york being the key and the value being a dataframe containing NYC's data. The same is true for Mumbai and Paris. As we grouped our data by city, we now have three different groups corresponding to those threee cities in the df.

###### To access your groupby data
To see the above picture in pandas...

```python
for city, city_df in g:
    print("city:", city)
    print("\n")
    print("data:", city_df)
    print("\n")
```

As you can see, the actual data in pandas groupby object looks very much like the pic

To access a specific dataframe...

```python
g.get_group('mumbai') # This allows us to see just the Mumbai data
```

###### Q1. What is the maximum temperature in each City?

```python
g.max() # Shows us the maximum temp in each city
# g is our df object that we created above
```

###### Split, Apply, Combine
What we have done is firstly divide our dataset into different groups based on City, then we run analytics on each of these groups and finally, we are combining the result into a single dataframe

###### Pictorial representation of the process looks like this...

```python
Image(filename = 'D:\\Pandas\\CodeBasics\\pics\\07splitapplycombine.jpg', height = 900, width = 900)
```

###### Q2. What is the average windspeed in each City?

```python
g.mean() # Will calculate the mean/average for all of our numeric columns
```

```python
g.describe() # Gives you a lot of analytics in one command
```

```python
%matplotlib inline
g.plot() # Not very clear but the first plot is for Mumbai, New York & finally Paris
```
