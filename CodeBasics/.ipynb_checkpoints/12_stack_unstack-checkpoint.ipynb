#### Stack()/Unstack()
This is another Pandas data reshaping utility

```python
import pandas as pd
from IPython.display import Image
df_stack = pd.read_excel('D:\\Pandas\\CodeBasics\\datasets\\12_stocks.xlsx', header = [0,1])
df_stack
```

This is our df. We had to supply the header information as we have two dolumnd that we want to use as the header.

To stack the inner most level of column headers, ie the company names in this example, we just need to call the stack function on the df...

```python
df_stack.stack()
```

We can see that it has taken the name of the comapnies from the columns and transpose them into rows.

If we wanted to take the first level of column headers instead, then we have to pass it the level parameter...

```python
df_stack.stack(level = 0)
```

Now we can see that both the price and the P/E information have become rows.

For more information on the stack() function, visit https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.stack.html

#### Unstack()
We use unstack() to perform the reverse of the stack() operation...

```python
df_stack02 = df_stack.stack()
```

```python
df_stack02 # This is the df that we are going to be working with
```

```python
df_stack02.unstack() # Gives us our original df back
```

You can perform stacking/unstacking with any number of column levels. Here we have a dataset with three levels of header data...

```python
Image(filename = 'D:\\Pandas\\CodeBasics\\pics\\levels.jpg')
```

```python
df_unstack = pd.read_excel('D:\\Pandas\\CodeBasics\\datasets\\12_stocks_3_levels.xlsx', header = [0,1,2])
df_unstack
```

We have to supply three values for the header this time as we have three levels of header information.

Again, if we call stack(), it will take the inner most level, ie the names of the companies, and transpose it...

```python
df_unstack.stack()
```

...and we see that the company information is now in the rows rather than the columns. Using the level argument, we can pick any level, ie 0, 1, or 2, to transpose...

```python
df_unstack.stack(level = 1)
```

...and we have the second row of header information moving down to the rows from the columns
