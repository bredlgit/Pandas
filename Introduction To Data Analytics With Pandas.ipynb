#### Let's add some rows at midnight on every day.
We want to get one value per day, to do this we are going to add one row, everyday at midnight and finding the value of the number of coffees at that time and then dropping my rows that are not at midnight.

We are going to create a new index but this one will live outside of the df(!?)

```python
# pd.date_range, with daily frequency, and normalisation
midnights = pd.date_range(data.index[0], data.index[-1], freq="D", normalize=True)

midnights
```

``pd.date_range()`` - We are going to use the date range method to create a new date range<br>
``data.index[0]`` - It is going to start at the first value in the index<br>
``data.index[-1]`` - It is going to end at the last entry in the index<br>
``freq="D"`` - This sets the frequency to days
        - We could have entries for weeks, 'W', months 'M' etc etc etc<br>
``normalize=True`` - Make these datapoints at midnight<br><br>

More information about the Pandas date_range() function can be accessed [here.](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.date_range.html)

These datetime objects no longer have a time, there ius just one a day, everyday, from the start of the df everyday to the end.

So we have this index of days and in our df we have an index that is randomly spaced. There are some indices that have uniformly spaced points which are all very neat but, at the moment, our index does not look like that...that's what our new index is for.

We are going to create a new index that takes the union of both of these.

#### Let's take the union of this index and our dataset's index.

```python
# Take the union of the existing and new indices
new_index = midnights.union(data.index)

new_index
```

``midgnights`` - This is the index that we created for one value at midnight everyday. It lives outside of the df<br>
``.union()`` - Join our two two values together<br>
``data.index`` - Is the current index of our df

This code instructs Python/Pandas to come up with a new index that has the values of both of these indices together in one index  and make sure that it is sorted in order<br>
**Note :** the union of these indices is just a new index where entries from both indices are present. It's sorted by time.

#### Now we can re-index our dataframe with this new index

```python
# .reindex() the dataframe
upsampled_data = data.reindex(new_index)

upsampled_data.head(10)
```

We can see that now that the two indices have been merged we have no data for the midnight index entries. Pandas doesn't fill in any data for us, it just fills in the missing data with 'I don't knows' or NaN's. For all the NaN's in the coffee column, we want them to have a value, it will be somewhere between the entries for that day and the midnight entry for the following day eg the first value that we have to fill in is something between 08:22 on 03/10/2011 and midnight on 04/10/2011. 

**Note :** `.reindex()` - keeps any values that conform to the new index, and inserts `NaN`s where we have no values. More information on the ``reindex()`` function can be found [at](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.reindex.html)

#### We can fill in these `NaN`s using interpolation
We could use ``.fillna('any_value')`` to set all the ``NaN``'s to a specific value but this is rather a blunt instrument and doesn't really help us. We could also have used the ``.fillna(ffill)`` & ``.fillna(bfill)`` methods but while these are a bit more subtle than they do not really help us either.

Interpolation - If I have a number of points and then one point that I do not know what its value is, we can look at the points to the left and right and...
- Linear Interpolation - We just take an average of the two points and make our unknown value that average
    - This is a pretty simple example of interpolation
    - This is ok if I am equally spaced between all of my points but our df index is not like that
- Interpolation Against Time - ie plot this against the time axis and, wherever the line is, take that value instead
    - This will be between the value befor and the value after
    - ...but it might be closer to the value after if the time that I'm sampling is closer to the value after

```python
# .interpolate the upsampled_data using the time method
upsampled_data = upsampled_data.interpolate(method="time")

upsampled_data.head(10)
```

- All the midnight values have now been filled in by pandas using the interpolate method.
- We do get some odd numbers eg 404.408 but this is the number interpolated between the time before and after by Pandas
    - This number is actually higher than the average of the two numbers either side of it
    - That's because midnight is closer to the following value than it is to the previous value therefore, the number is closer to the value following

#### We're now ready to resample the time-series at a daily frequency.

```python
# .resample() followed by .asfreq()
daily_data = upsampled_data.resample("D")
#'D' is for daily
daily_data
```

This is a DateTimeIndexResampler & it actually looks like ``groupby()`` object. Like a ``groupby()`` object it mashes some things together but it doesn't really gives us any information<br>
This says, look at the timestamps and group everything into days, if n rows are in the same day then put them in the same group but we haven't told it what to do with the group yet. We could do several things...<br>
- We could take the mean of all the bins(?) in the group...
    - ``daily_data = upsampled_data.resample("D").mean()``
- I could take the count of everything in the group...
    - ``daily_data = upsampled_data.resample("D").count()``
- What we are actually gonna do is to say asfreq()...
    - Which basically says, resample daily and only take the midnights from that day
    - As we have inferred the values at those midnights, we get a number for the whole group at midnight

```python
daily_data = upsampled_data.resample("D").asfreq()
daily_data
#You can use the on argument to resample on a column
```

We have lost the time information in our index but we don't need it anymore because everything is equally spaced one day apart

We still have a weekdays and a contributors column which we do not need anymore...the reason that we do not need them is that they are full of NaNs, they do not contain any information. However, we are only going to drop the contributor column...

```python
# Drop the contributor column, we no longer need it
daily_data = daily_data.drop(["contributor"], axis=1)
daily_data.head()
```

We can see that the contributor column has been dropped

```python
# Generate a column of weekday names
daily_data["weekdays"] = daily_data.index.weekday_name  # We did it the slow way before...
daily_data.head()
```

This is the easier way to get the weekdays into our df, see above for the more difficult way. 

What we have done is taken the weekdays column and set it as the ``df.index.weekday_name``. You take the index, which is a list of dates and you say what's the day of the week for each of those and so we have that information back.

What we have ended up with is our original df back but instead of being randomly set with times and dates, it is now uniformly homogenously sampled, one a day at midnight.

```python
# Bring in data/department_members.csv;
# have the first column be the index, and parse the dates
people = pd.read_csv("data/department_members.csv", index_col=[0], parse_dates=True)

people.head() # At the end of August there were 12 people etc
```

#### Let's join the datasets.
How do we join something that's monthly to our dataset that's daily? Firstly, we are going to do an outer join.<br>
**How join works**
- A join looks at two datasets, it's going to compare one of the columns that you are telling it to look at, usually that'll be the index but you can tell it to look at anything. 
- You then say, wherever these indices match extend that row
- The default in Pandas is an inner join
    - If my first dataset has an index which doesn't appear in my second dataset - drop that row!
- An outer join does the opposite
    - If this one has an index and this one hasn't, keep that row and just fill it with NaN's

**Note :** by default, inner joins are performed. That is, if a row from one of the datasets has an index that isn't in the other dataset, that row is dropped. You can specify whether you want outer, left, or right joins, as well plenty of other useful options. The `pandas` API for joining or [merging](https://pandas.pydata.org/pandas-docs/stable/merging.html) datasets is very developed.

```python
#This is what the outer join looks like...
daily_data.join(people, how="outer")
```

Let's take the 31/08/2011 entry to explain what join has done here. The daily_data df has no data for this date so we see a row of NaN's from that dataset. The people df says that there were 12 members of the dept. So although we do not start recording the number of coffees until 3/10/11 (from daily_data), we see that the other df is supplying data. Hence we have 3 x NaN and one piece of useful data<br>
For more information, have a look at the Pandas' docs on [Merge, Join & Concatenate](https://pandas.pydata.org/pandas-docs/stable/merging.html)

Now we are going to interpolate over the members column because we have a bunch of NaNs which we don't like...we wanna fill those in...

```python
# Use an outer join, then interpolate over missing values using nearest values
daily_data = daily_data.join(people, how="outer").interpolate(method="nearest")

daily_data.head()
```

#### Let's create a column for the number of coffees consumed per person

```python
# New column is the ratio of coffees made on a given day to number of members in the department
daily_data["coffees_per_person"] = daily_data.coffees_made_today / daily_data.members

# Let's drop those remaining NaNs while we're at it
#daily_data.dropna(inplace=True)

daily_data.head()
```

This is not the number of coffees per person! This is just an avaerage given the data and the number of people in the dept., you may well have people who do not drink any coffee at all and people who drink more than they should...it is an average not a true figure

#### We can now plot this column

```python
# Plot the coffees_per_person column
daily_data.coffees_per_person.plot()
```

- The huge up and downs are being caused by the drop in consumption over the weekends...it's hard to see but that's what it is
- The plateaus at the bottom is where the machine was broken and didn't make any coffees at all!
