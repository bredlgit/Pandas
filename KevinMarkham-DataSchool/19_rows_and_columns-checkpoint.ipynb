```python
import pandas as pd
ufo = pd.read_csv('http://bit.ly/uforeports')
ufo.head()
```

```python
ufo.loc[0, :] # Using loc to retrieve row zero and all columns in that row
```

```python
ufo.loc[[0, 1, 2],  :] # to retrieve rows 0, 1 & 2 and all their columns
```

```python
ufo.loc[0:2,  :] # Gives us the same result as above
```

```python
ufo.loc[:, 'City'] # To select all rows but just the City column
```

```python
ufo.loc[:, ['City', 'State']] # This retrieves both City and State column 
```

```python
ufo.loc[:, 'City':'State'] # Specified as a range - City through State
```

```python
ufo.loc[0:2, 'City':'State'] # We can combine this to select both rows and columns
```

```python
ufo.head(3).drop('Time', axis=1)
```

```python
ufo[ufo.City == 'Oakland'] # To select all rows where City equals Oakland
```

```python
ufo.loc[ufo.City == 'Oakland', :] # Gives us the same result as above
```

```python
ufo[ufo.City == 'Oakland'].State # This is called chained indexing and can cause problems
```

```python
ufo.loc[ufo.City == 'Oakland', 'State']
```

```python
ufo.iloc[:, [0, 3]]
```

```python
ufo.iloc[:, 0:4] # We can also use a range here as well
```

```python
ufo.iloc[0:3, :] # to retrieve rows 0, 1 & 2 and all columns
```

```python
ufo[['City', 'State']] # Shortcut for selkecting multiple columns
```

```python
ufo[0:2] # This is referring to rows by integer position
```
