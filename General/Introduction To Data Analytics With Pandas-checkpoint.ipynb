### Cleaning the data
**Problem One - The coffees column should only contain numerical data.**

```python
# cast the coffees column using pd.to_numeric, and coerce errors
data.coffees = pd.to_numeric(data.coffees, errors = 'coerce')

data.head()
```

If we tell Pandas to convert our column to numeric values only it will have a good go. (We use to_numeric first as it handles errors quite well.) The code above throws an error because we have a string, testing, in there and it cannot parse that. NaN values are ignored as they are actually valid elements<br>
With *<font color=blue>errors = 'coerce'</font>* added to the command, we can tell Pandas to turn any errors into an NaN. The testing string is the only error that we have and it is safe to discard it as it doesn't provide any useful information...

```python
data.dtypes
```

We can see the to_numeric argument ran fine, we can see our df and when we check the data types of the different columns, we see that columns is now a float. We have some NaN's but we were expecting those.<br><br>

**On which weekdays were contributions made?**<br>
Because the timestamp column is a datetime column, it opens up a whole new set of functions that we can use that are all date related. The one that we are using here is dt, or datetime, and we are using that to determine the days of the week that the contributions were made

```python
# Create a series of the weekdays for each entry using .dt.weekday
weekdays = data.timestamp.dt.weekday
weekdays # Shows us the days of the week that coffee was made in the machine with zero being Monday
```

```python
# assign() it to our dataframe
data = data.assign(weekdays=weekdays)
#The weekdays column is going to be the weekdays object that we just created
data.head()
```

**df.assign - New way to create a new column in the df**<br>
For more information, visit - https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.assign.html<br>
**Note:** df.assign takes a copy, rather than a view, and adds the new column to the df. This may or may not cause issues later on.<br>
For more information, visit - http://pandas.pydata.org/pandas-docs/stable/indexing.html#indexing-view-versus-copy

**Can we replace these integers with actual weekdays?**
While it is great to be able to pull out the 0-6 notation, it is not very helpful when it comes to days of the week especially as there is a much simpler way of recording these. A way that everyone understands...

```python
weekday_names = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
weekday_dict = {key: weekday_names[key] for key in range(7)}
```

 - weekday_names - A list of the days of the week
 - weekday_dict - A mapping of the days of the week to the numbers 0-6 (range(7))
 - Then we iterate through them

Use .apply() to apply a custom function to the weekdays column...This code was on the github...
*<font color = blue>data.weekdays = data.weekdays.apply(lambda x: weekday_dict[x])</font>* but, in the video, he actually used something completely different.<br>
**lambda functions** - Here is an anonymous function, don't bother giving it a name just apply it<br>
For more information, visit - https://www.programiz.com/python-programming/anonymous-function

**What did he actually use**

```python
def day_of_week(idx):
    return weekday_dict[idx]
day_of_week(0)
```

```python
day_of_week(2)
```

First we defined our function and made sure that it worked and brough back the expected results. Then we applied it to the weekdays column/Series of our data df...

```python
data.weekdays = data.weekdays.apply(day_of_week)
data.head()
```

Now, we see that our weekdays columns actually returns the days of the week in a format that we can all understand<br><br>
**Let's group by these weekdays.**<br>
The original question was, on which days are people contributing...

```python
# .groupby() the weekdays and then .count() rows in each group
weekday_counts = data.groupby("weekdays").count()

weekday_counts
```

We are taking the df and using .groupby at the weekdays column to group those entries into the Monday (etc) groups and then count them up<br>
What .groupby() does is that it takes all the Mondays together and counts them, and it will do that for every unique entry in weekdays column(when we point it at that column).<br><br>
**Note:** this first line could be replaced by weekday_counts = data.weekdays.value_counts(), with the only difference being that that would return a series to us, and here, we got back a dataframe.

```python
# We can reorder this dataframe by our weekday_names list
weekday_counts = weekday_counts.loc[weekday_names]
weekday_counts
```

**We can now visualise these weekday counts**

```python
# .plot() a bar chart of data in weekday_counts
weekday_counts.timestamp.plot(kind="bar", title="Datapoints added on each weekday")
```

The original code looked like this. This is because we have three columns which all show the same information. As we do not need the same information represented three times, we picked a column to plot, it really doesn't matter which one as they all have the same data, and we got the ploty above.
